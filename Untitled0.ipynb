```
from torchvision import *
from torch import *  
import numpy as np
import csv
from tkinter import *

#code for mnist handcharacter recognition, has a weights class and a model class
#and an application class, the application class is part of tkinter, to create 
#the user interface...

#the model creates an individual network out of the image, 
#which is really a csv file that contains intensity values for the characters
#it uses MNIST on csv dataset. 
#It also does forward and backward propagartion with its functions
#The weights class creates the weights matrix for
#the current network...which will be the B matrix , containing the b values of 
#the current image, the A matrix as well as the Theta bias matrix 


# the applicaltion class handles the user interface which has a button to
#innitiate training and one to innitiate testing.

#I would like to adapt this to a cancer detection model. differences will be 
# that we will have a B matrix for each of the RGB channels as well as a weight 
#matrix. Also we will be dealing with real images and not a csv.

#Also we need to do the hamadard product between the pos matrix and each of the 
#B and A matrices before we multiply B and A together. that has not been coded 
#for quite yet


lr=0.01
B = randn(255,requires_grad=True)  #B,A,Theta values of all the neurons are 
                                   #instantiated 
A = randn(255,requires_grad=True)
theta=randn(255,requires_grad=True)

#Target is the matrix containg the valsues of the traget image, with values
# 1 to 10 for all training examples

target=randn(255,requires_grad=True)

#pos is a randomly innitialised matrix.  pos[0]  does a hamardad product with 
#the a and b values in the first position of the image
#and so on. these values will also be changed throgh back prop and help the 
#system know which pixel value is where.

pos=randn(255,requires_grad=True)

#outx is the output layer which has 10 neurons for each of the 10 numbers.

outx=randn(9,requires_grad=True)

#weights class for creating the weight matrix for b , a and theta (bias) values
#we may optionally just keep theta as 1 as is common

class weights():
    def __innit__ (self):#############
        
        #defines the B,A and theta matrices for current network
        self.weightsB=self.BB
        self.weightsA=self.AA
        self.weightstheta=self.TT
        
        #creates B matrix
    def BB(self):
        BB= torch.FloatTensor()
        
        ifile=open(mnist_train.txt,"rU")
        reader = csv.reader(ifile,delimiter=",")
        numcol=0
        for row in reader:
            if numcol==0:
                cat(target,row,0)
                numcol+=1
                continue
            elif numcol != 0:
            
                cat(BB,B[row],0)
                
            numcol += 1
        return BB
    
         #creates A matrix
    def AA(self):
        AA=FloatTensor()
        
        ifile=open(mnist_train.txt,"rU")
        reader = csv.reader(ifile,delimiter=",")
        numcol=0
        for row in reader:
            if numcol==0:
                
                numcol+=1
                continue
            elif numcol != 0:
                cat(AA,A[row],0)       
                
                
                numcol += 1
        return AA
        #creates bias matrix
    def TT(self):
       
       TT=FloatTensor()
       ifile=open(mnist_train.txt,"rU")
       reader = csv.reader(ifile,delimiter=",")
       numcol=0
       for row in reader:
            if numcol==0:
               numcol+=1
               continue
            elif numcol != 0:
                cat(theta,theta[row],0)
                
               
                numcol += 1
         
            return TT
    #Model class is responsible for forward prop and back prop.
class model():

    def __innit__(self):
        pass
    
    #forward propagation
    def forw(self,w):
        x=self.weightsB*pos
        y=self.weightsA*pos
        z=self.weightstheta*pos
        n=x*y+z
        self.tanh1=self.tanh(n) 
        s=self.tanh1
        o1w=outx[0]*s
        o2w=outx[1]*s
        o3w=outx[2]*s
        o4w=outx[3]*s
        o5w=outx[4]*s
        o6w=outx[5]*s
        o7w=outx[6]*s
        o8w=outx[7]*s
        o9w=outx[8]*s
        o10w=outx[9]*s
        soft=tensor[o1w,o2w,o3w,o4w,o5w,o6w,o7w,o8w,o9w,o10w]
        return softmax(soft)
    #activation functions 
    def tanh(self,x):
        self.tanh=nn.tanh(x)
        return self.tanh
    
    def softmax(self,w,t=1):
        self.soft=nn.softmax(w)
        return self.soft
    
    #backprop
    def back(self,w):
        self.error =100
        while self.error > 0.1:
            
            
                       
            self.s=self.forw(w)
            self.criterion=nn.CrossEntropyLoss()
            self.loss =criterion(self.s)
            self.error = self.loss
            self.loss.backward
            self.A()
            self.B()
            self.T()
            
            self.displayresult(self.error)
            
        #save weights values after training each image    
        save(w,'Weights') 
        pass
        
        # testing function activated when test button is clicked
        
    def testing(self,w):
         
                  
             
         self.s=self.forw(w)
         self.criterion=nn.CrossEntropyLoss()
         self.loss=self.criterion(self.s)
          
            
         
    
        
         return self.loss
        
        
           
      #update of weight values  
               
    def A(self):
        self.weightsA=self.weightsA-lr*self.weightsA.grad
        return self.weightsA
    def B(self):
        self.weightsB=self.weightsB-lr*self.weightsB.grad
        return self.weightsB
    def T(self):
        self.weightsT=self.weightsT-lr*self.weightsT.grad
        return self.weightsT
    
    # application class for user interface using tkinter
    
    
class application(Frame):
    def createwidgets(self):
        self.trains=Button(self)
        self.trains["text"]= "Train"
        self.trains["command"]=running1()

        self.trains.pack({"side":"left"})
#### ########
        self.test=Button(self)
        self.test["text"]= "test"
        self.test["command"]=testing1()

        self.test.pack({"side":"left"})
        self.display = Text(root)
      
        
        self.display.pack({"side":"left"})
        self.displaytest = Text(root)
      
        
        self.displaytest.pack({"side":"left"})
        pass
    
    # This runns the training process
    def runningl(self):
        i=0
        ifile=open(mnist_train.txt,"rU")
        reader = csv.reader(ifile,delimiter=",")
        for row in reader:
            model[i]=model()
            weights[i]=weights()
            model[i].back(weights[i])
            i+=1
    pass
     #runs testing process.       
    def testing1(self):
        
        i=0
        self.totalerror=0
        ifile=open(mnist_test.txt,"rU")
        reader=csv.reader(ifile,delimiter=",")######
        for row in reader:
            model[i]=model()
            weights=load('Weights')
            self.error[i]=model[i].testing(weights,i)
            self.totalerror+=self.error[i]
            i+=1
            
        self.Neterror=(self.totalerror/i) *100
        self.displaytest.insert(self.Neterror)
    pass
            
    def __init__(self,master=None):
        Frame.__init__(self,master)
        self.pack()
        self.createwidgets

root=Tk()
app=application(master=None)
app.mainloop()
root.destroy()


```
